## Problem 1

1. $Y = sin(U) $, where $ U \sim \text{Uniform}(0, \pi/2) $

2. **PDF of $ U $**: $ f_U(u) = \frac{2}{\pi} $ for $ u \in [0, \pi/2] $.

3. **Inverse Transformation and its Derivative**: The inverse transformation is $ u = \sin^{-1}(y) $ with the derivative $ \frac{du}{dy} = \frac{1}{\sqrt{1 - y^2}} $.

4. **Applying the Transformation of Variables**: The PDF of $ Y $ is obtained by applying the transformation of variables formula:
   $ f_Y(y) = f_U(\sin^{-1}(y)) \cdot \left| \frac{du}{dy} \right| = \frac{2}{\pi} \cdot \frac{1}{\sqrt{1 - y^2}}, \quad y \in [0, 1] $

## Problem 2

### Task 1

1. **Condition for Real Roots**: For $ t^2 + tX + Y = 0 $, real roots exist if $ \Delta = X^2 - 4Y \geq 0 $.

2. **Joint PDF of $ X $ and $ Y $**: Since $ X $ and $ Y $ are i.i.d. uniform on $(-b, b)$, their joint PDF is $ f_{X,Y}(x, y) = \frac{1}{2b} \cdot \frac{1}{2b} $.

3. **Probability Calculation**: 
   $ P(X^2 \geq 4Y) = \int_{-b}^{b} \int_{-\frac{x^2}{4}}^{\frac{x^2}{4}} \frac{1}{2b} \cdot \frac{1}{2b} \, dy \, dx $

**Answer**: $ P(X^2 \geq 4Y) = \frac{b}{12} $

### Task 2

```python
import matplotlib.pyplot as plt
import numpy as np

def probability(b):
    return b / 12

b_values = np.linspace(0, 10, 100) 
probabilities = probability(b_values)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(b_values, probabilities, label='P(X^2 >= 4Y)')
plt.xlabel('b')
plt.ylabel('Probability')
plt.title('Probability of Real Roots as a Function of b')
plt.legend()
plt.grid(True)
plt.show()
```
