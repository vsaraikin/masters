## Problem 1

Consider a set $S$ with $m + n$ distinct elements, where $m$ and $n$ are non-negative integers. The number of ways to choose $k$ elements from this set is given by the binomial coefficient $\binom{m+n}{k}$.

Let's divide the set $S$ into two disjoint subsets, $A$ and $B$, where $A$ has $n$ elements and $B$ has $m$ elements. Let's can form a subset of $k$ elements from $S$ by choosing $j$ elements from $A$ and $k-j$ elements from $B$, for some $j$ between 0 and $k$. The number of ways to choose $j$ elements from $A$ is $\binom{n}{j}$, and the number of ways to choose $k-j$ elements from $B$ is $\binom{m}{k-j}$.

Since $j$ can range from 0 to $k$, sum over all possible values of $j$ to count all the ways of forming a $k$-element subset from $S$. This gives us the sum:
$$\sum_{j=0}^{k} \binom{n}{j} \binom{m}{k-j}$$

Each term in this sum counts the number of $k$-element subsets of $S$ formed by taking $j$ elements from $A$ and $k-j$ elements from $B$. Since this counts all possible $k$-element subsets of $S$, the sum must equal the total number of $k$-element subsets of $S$, which is $\binom{m+n}{k}$ $=>$ $$\sum_{j=0}^{k} \binom{n}{j} \binom{m}{k-j} = \binom{m+n}{k}$$

## Problem 2

Given:
- Probability of drawing a white ball, $ p = \frac{m_1}{m} $
- Probability of drawing a black ball, $ q = 1 - p = \frac{m_2}{m} $

Since the balls are drawn with replacement $=>$ each draw is independent.

The probability of drawing exactly $ r $ white balls in $ n $ trials is given by:

$$
P(X = r) = \binom{n}{r} p^r q^{n-r}
$$

where $ \binom{n}{r} $ is the binomial coefficient, calculated as $ \frac{n!}{r!(n-r)!} $. This coefficient represents the number of different ways to choose $ r $ successes (white balls) out of $ n $ trials.

Answer: Binomial Distribution

## Problem 3

Given $ X \sim \text{Bi}(n, p) $ and $ Y \sim \text{Bi}(m, p) $ are independent binomial random variables $=>$ the probability mass function (PMF) of $ X $ is given by:
$$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} $$
for $ k = 0, 1, 2, \ldots, n $.

The conditional probability $ P(X = k | X + Y = r) $ is:
$$ P(X = k | X + Y = r) = \frac{P(X = k \, \text{and} \, X + Y = r)}{P(X + Y = r)} $$

Since $ X $ and $ Y $ are independent, $ P(X = k \, \text{and} \, X + Y = r) $ simplifies to $ P(X = k)P(Y = r - k) $.

The probability $ P(X + Y = r) $ is the sum of the probabilities of all ways in which $ X $ and $ Y $ can add up to $ r $, considering their respective binomial distributions. This is given by:
$$ P(X + Y = r) = \sum_{k=0}^{r} P(X = k)P(Y = r - k) $$

Substituting:
$$ P(X = k | X + Y = r) = \frac{\binom{n}{k} p^k (1-p)^{n-k} \binom{m}{r-k} p^{r-k} (1-p)^{m-(r-k)}}{\sum_{k=0}^{r} \binom{n}{k} p^k (1-p)^{n-k} \binom{m}{r-k} p^{r-k} (1-p)^{m-(r-k)}} $$

**Answer:** hypergeometric distribution.


## Problem 4

### 1

- Player A chooses a number $ p_1 $ between 0 and 1
- Player B, knowing $ p_1 $, must choose $ p_2 $ to maximize their winning chances
- The outcomes are:
    - With probability $ p_1 $, $ p_1 $ becomes 0
    - With probability $ p_2 $, $ p_2 $ becomes 0
    - The player with the non-zero number wins, or it's a draw if both are zero.

<br>

- Strategy for Player Two:
    - If $ p_1 = 0 $, choose any $ p_2 > 0 $ to win
    - If $ p_1 = 1 $, choose 1 for a draw
    - For $ 0 < p_1 < 1 $, choose $ p_2 = p_1 $. This leads to a draw or a win for Player Two if $ p_1 $ becomes zero but $ p_2 $ does not.


### 2

- Player One must choose a number to maximize their chance of winning or drawing
- Strategy for Player One:
    - Choosing any $ p_1 $ where $ 0 < p_1 < 1 $ will most likely end in a draw, as Player Two will choose $ p_2 = p_1 $
    - Choosing $ p_1 = 0 $ guarantees a loss
    - Choosing $ p_1 = 1 $ leads to a draw, as Player Two will also choose 1

- Therefore, the optimal strategy for Player One is to choose $ p_1 = 1 $ to ensure at least a draw.


## Problem 5

Let $ X $ be the random variable representing the number of failures before achieving $ r $ successes.

$ P(X = k) $, is the probability that there are $ k $ failures before the $ r $-th success.


- The probability of $ r - 1 $ successes and $ k $ failures in $ r + k - 1 $ trials is given by the binomial coefficient $ \binom{r + k - 1}{r - 1} $ multiplied by $ p^{r-1}(1-p)^k $.
- The probability of the $ r $-th success on the $ (r + k) $-th trial is $ p $.
- Therefore, the PMF is given by the product of these probabilities:
$ P(X = k) = \binom{r + k - 1}{r - 1} p^{r-1}(1-p)^k \cdot p $
- Simplifying:
$ P(X = k) = \binom{r + k - 1}{k} p^r (1-p)^k $ for $ k = 0, 1, 2, \ldots $

**Answer:** negative binomial distribution


## Problem 6

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import nbinom

r_values = [5, 10, 20]  # Number of successes
p_values = [0.2, 0.5, 0.8]  # Probability of success
x = np.arange(0, 100)  # Range of x values (number of failures)

# Plotting
for r in r_values:
    plt.figure(figsize=(12, 8))
    for p in p_values:
        pmf = nbinom.pmf(x, r, p)
        plt.plot(x, pmf, label=f'p = {p}')
    
    plt.title(f'Negative Binomial PMF with r = {r}')
    plt.xlabel('Number of Failures (k)')
    plt.ylabel('PMF')
    plt.legend()
    plt.grid(True)
    plt.show()
```
