## Problem 1

```python
# Define the range of x (from 0 to 5 for good visibility)
x = np.linspace(0, 5, 100)

# PDF of Exponential Distribution (λ = 1)
pdf = expon.pdf(x)

# CDF of Exponential Distribution (λ = 1)
cdf = expon.cdf(x)

# Plotting both PDF and CDF
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label='PDF (λ = 1)')
plt.plot(x, cdf, label='CDF (λ = 1)')
plt.title('Exponential Distribution (λ = 1)')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## Problem 2

Consider an exponential distribution with parameter $ \lambda $. The survival function (complement of the CDF) is given by $ S(x) = e^{-\lambda x} $ for $ x \geq 0 $.

To prove the memoryless property, let's show that for any $ s, t \geq 0 $, the following holds:

$ P(X \geq s + t | X \geq s) = P(X \geq t) $

Using the definition of conditional probability and the survival function:

$ P(X \geq s + t | X \geq s) = \frac{P(X \geq s + t)}{P(X \geq s)} = \frac{e^{-\lambda (s + t)}}{e^{-\lambda s}} $

$ P(X \geq s + t | X \geq s) = e^{-\lambda t} = S(t) = P(X \geq t) $

Therefore, the exponential distribution satisfies the memoryless property.

## Problem 4

### **CDF of the Minimum of Two Exponential Random Variables**

Consider independent random variables $ X \sim \text{Exp}(\lambda_X) $ and $ Y \sim \text{Exp}(\lambda_Y) $. The cumulative distribution function (CDF) of $ Z = \min\{X, Y\} $ is derived as follows:
\begin{align*}
    F_Z(z) &= P(Z \leq z) = 1 - P(X > z, Y > z) \\
           &= 1 - P(X > z) \cdot P(Y > z) \\
           &= 1 - e^{-\lambda_X z} \cdot e^{-\lambda_Y z} \\
           &= 1 - e^{-(\lambda_X + \lambda_Y) z}
\end{align*}

### **CDF of the Sum of Identical Exponential Random Variables**

When $ \lambda_X = \lambda_Y = \lambda $, the sum $ Z = X + Y $ follows a Gamma distribution. The probability density function (PDF) of $ Z $ is obtained by the convolution of the PDFs of $ X $ and $ Y $:
\begin{align*}
    f_Z(z) &= \int_{0}^{z} f_X(x) f_Y(z - x) dx \\
           &= \lambda^2 \int_{0}^{z} e^{-\lambda x} e^{-\lambda (z - x)} dx \\
           &= \lambda^2 z e^{-\lambda z}
\end{align*}
The CDF of $ Z $, $ F_Z(z) $, is the integral of $ f_Z(z) $:
\begin{align*}
    F_Z(z) &= \int_{0}^{z} \lambda^2 t e^{-\lambda t} dt
\end{align*}
This integral represents the CDF of a Gamma distribution with parameters $ k = 2 $ and $ \theta = \frac{1}{\lambda} $.

\begin{align*}

           &= e^{-e^{-y}}
\end{align*}
